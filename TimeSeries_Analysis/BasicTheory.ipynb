### 9장. 시계열을 위한 머신러닝

시계열 분석에 머신러닝을 적용

과정에 규칙이 존재하거나, 특정 형식의 과정 자체가 존재한다는 상정을 하지 않는 방법론. 우리가 원하는 결과 예측에 관련된 과정의 행동을 묘사할 수 있는 패턴을 식별하는 것에 집중. 시계열 클러스터링 형식의 시계열에 대한 비지도학습 고려.

클러스터링과 트리 기반 방법론으로 예측과 분류 문제를 다룬다. 시계열의 특징 생성은 트리 기반 방법론에서 필수적, ARIMA 모델과는 달리 '시간을 인식'하지 못하기 때문.

클러스터링 및 거리 기반의 분류는 '입력'으로 시계열 특징 생성 가능, 입력을 하려면 동적시간워핑(DTW) 이라는 거리평가지표를 알아야 한다.

동적시간워핑(DTW) : 데이터 전체에 대한 시간 정보 보존

```python
# 시계열 특징 만들기 예시
! pip install cesium
from cesium import datasets
egg = datasets.fetch_andrzejak()
  # [dataset 불러오기]
```

```python
# 시계열 시각화

# 첫 번째 plot
plt.subplot(3,1,1)
plt.plot(egg["measurements"][0])
plt.legend(egg['classes'][0])
  # [legend : 범례지정]

# 두 번째 plot
plt.subplot(3,1,2)
plt.plot(egg["measurements"][300])
plt.legend(egg["classes"][300])

# 세 번째 plot
plt.subplot(3,1,3)
plt.plot(egg["measurements"][450])
plt.legend(egg["classes"][450])

  # [진폭이 유용할 수 있다는 것을 시사]
```
